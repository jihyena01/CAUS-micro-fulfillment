```python
import pandas as pd

# 파일 경로
file_path = ''

# CSV 파일 읽기
df = pd.read_csv(file_path)

# 데이터 크기 및 컬럼 확인
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("Columns:", df.columns.tolist())

# 상위 5개 행 확인
df.head()
```

```python
depot_candidates = df[['senderAddress', 'senderLat', 'senderLng']].drop_duplicates()
print(f"Unique depots: {len(depot_candidates)}")
depot_candidates.head()
```

```python
# Customer별 수요 집계
customer_demand = df.groupby(['receiverAddress', 'receiverLat', 'receiverLng'])['weight'].sum().reset_index()
customer_demand.columns = ['Address', 'Latitude', 'Longitude', 'TotalWeight']

# 수요를 10배 스케일링 후 정수로 변환
customer_demand['TotalWeight'] = (customer_demand['TotalWeight'] * 10).round().astype(int)

print(f"Unique Customers: {len(customer_demand)}")
customer_demand.head()
```

```python
import folium
import numpy as np

np.random.seed(42)
depot_sample = depot_candidates.sample(10, random_state=42)
customer_sample = customer_demand.sample(200, random_state=42)

# 지도 생성
m = folium.Map(location=[10.78, 106.7], zoom_start=12)

# Depot 마커: 빨간 네모
for _, row in depot_sample.iterrows():
    folium.RegularPolygonMarker(
        location=[row['senderLat'], row['senderLng']],
        number_of_sides=4,  # 네모
        radius=6,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.8,
        popup=f"Depot: {row['senderAddress']}"
    ).add_to(m)

# Customer 마커: 파란 원형
for _, row in customer_sample.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=4,
        color='blue',
        fill=True,
        fill_opacity=0.5,
        popup=f"Customer: {row['Address']}, Demand: {row['TotalWeight']:.1f} kg"
    ).add_to(m)

# 지도 출력
m
```

```python
# 고객 좌표의 최소/최대값 계산
min_lat = customer_demand['Latitude'].min()
max_lat = customer_demand['Latitude'].max()

min_lng = customer_demand['Longitude'].min()
max_lng = customer_demand['Longitude'].max()

print(f"Latitude (Y) Range: min = {min_lat}, max = {max_lat}")
print(f"Longitude (X) Range: min = {min_lng}, max = {max_lng}")
```

```python
# 고객 수요 출력: 가장 적은 10개
print("\n[Customers with Lowest Demand (Bottom 10)]")
print(customer_sample[['Address', 'TotalWeight']].sort_values(by='TotalWeight').head(10).to_string(index=False))

# 고객 수요 출력: 가장 많은 10개
print("\n[Customers with Highest Demand (Top 10)]")
print(customer_sample[['Address', 'TotalWeight']].sort_values(by='TotalWeight', ascending=False).head(10).to_string(index=False))

# 총 수요량 출력
total_demand = customer_sample['TotalWeight'].sum()
print(f"\nTotal Demand (in sampled 200 customers): {total_demand:.2f} kg")
print(f"\nAvg Demand per Customer: {total_demand / len(customer_sample):.2f} kg")
```

### Instance Generator

```python
import os
import numpy as np

def generate_realistic_instance(depot_sample, customer_sample, output_folder, instance_index=1):
    C = len(customer_sample)
    D = len(depot_sample)

    # 좌표 추출
    depot_coords = list(zip(depot_sample['senderLat'], depot_sample['senderLng']))
    cust_coords = list(zip(customer_sample['Latitude'], customer_sample['Longitude']))

    # 수요 추출
    cust_demands = customer_sample['TotalWeight'].tolist()
    total_demand = sum(cust_demands)
    mean_demand = np.mean(cust_demands)

    # ✅ Vehicle Capacity:
    vehicle_cap = int(np.max(cust_demands))

    # ✅ Depot Capacity: vehicle_cap × multiplier (13~18 랜덤)
    cap_multiplier = np.random.randint(2, 4)  # 정수 [13,18] 포함
    depot_caps = [vehicle_cap * cap_multiplier for _ in range(D)]

    # ✅ Open Costs:
    open_costs = [np.random.randint(10, 20) for _ in range(D)]

    # ✅ 저장 경로 및 파일 이름
    os.makedirs(output_folder, exist_ok=True)
    filename = f"set200-10-{instance_index}_vc{vehicle_cap}.txt"
    path = os.path.join(output_folder, filename)

    # ✅ 파일 작성
    with open(path, 'w') as f:
        f.write(f"{C}\n{D}\n\n")

        # Depot coords
        for x, y in depot_coords:
            f.write(f"{x:.6f}\t{y:.6f}\n")
        f.write("\n")

        # Customer coords
        for x, y in cust_coords:
            f.write(f"{x:.6f}\t{y:.6f}\n")
        f.write("\n")

        # Vehicle capacity
        f.write(f"{vehicle_cap}\n\n")

        # Depot capacities
        for cap in depot_caps:
            f.write(f"{cap}\n")
        f.write("\n")

        # Customer demands
        for dem in cust_demands:
            f.write(f"{dem}\n")
        f.write("\n")

        # Depot opening costs
        for oc in open_costs:
            f.write(f"{oc}\n")
        f.write("\n")

        # Fixed values
        f.write("1000\n\n0\n\n")

    print(f"[Saved] {path}")
    
output_folder = "benchmark_instances_realdata"
generate_realistic_instance(depot_sample, customer_sample, output_folder, instance_index=1)
```

```python
import folium, matplotlib
from matplotlib import colors as mcolors

# ─────────────────────────────────────────────
# (0) 팔레트 정의 : 빨강·파랑 제외, 고채도 10색
# ─────────────────────────────────────────────
ROUTE_COLORS = [
    "#FF8C00",  # Vivid Orange
    "#9C27B0",  # Deep Purple
    "#00BCD4",  # Bright Cyan
    "#C2185B",  # Raspberry
    "#009688",  # Teal
    "#FFD600",  # Neon Yellow
    "#795548",  # Mocha Brown
    "#E91E63",  # Hot Pink
    "#7C4DFF",  # Electric Violet
    "#FF5722",  # Pumpkin
]

# ─────────────────────────────────────────────
# (1) 지도 준비
# ─────────────────────────────────────────────
m = folium.Map(location=[10.78, 106.7], zoom_start=12)

# depot : 진한 빨강 네모
for d, (lat, lng) in enumerate(depot_coords):
    folium.RegularPolygonMarker(
        location=(lat, lng), number_of_sides=4, radius=6,
        color="#FF0000", fill_color="#FF0000", fill_opacity=0.9,
        popup=f"Depot {d}"
    ).add_to(m)

# customer : 진한 파랑 원
for cid, (lat, lng) in enumerate(customer_coords):
    folium.CircleMarker(
        location=(lat, lng), radius=3,
        color="#0066FF", fill=True, fill_color="#0066FF",
        fill_opacity=0.7, popup=f"Cust {cid}"
    ).add_to(m)

# ─────────────────────────────────────────────
# (2) 경로 선 : 팔레트에서 겹치지 않는 색만 사용
# ─────────────────────────────────────────────
for dep_idx, dep_routes in enumerate(route_json["routes"]):
    for r_idx, rt in enumerate(dep_routes):
        ids     = list(map(int, rt["Ids"]))
        coords  = [
            depot_coords[ids[0]] if i in (0, len(ids)-1)
            else customer_coords[ids[i]]
            for i in range(len(ids))
        ]
        color   = ROUTE_COLORS[(dep_idx * 3 + r_idx) % len(ROUTE_COLORS)]

        folium.PolyLine(
            locations=coords, color=color,
            weight=3, opacity=0.9,
            tooltip=f"D{dep_idx}-R{r_idx}"
        ).add_to(m)

m.save("route_map.html")
print("🗺️ route_map.html (마커-색과 겹치지 않는 경로색) 완료")


m
```
